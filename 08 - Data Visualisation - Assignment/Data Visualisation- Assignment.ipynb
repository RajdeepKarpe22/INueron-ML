__In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.__

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris_dataset = datasets.load_iris()
iris_dataset
```

```python
X = iris_dataset['data']
y = iris_dataset['target']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X = pca.fit_transform(X)
X
```

```python
fig = plt.figure(1, figsize=(7, 6))
plt.clf()
axes = Axes3D(fig, rect=[0, 0, .95, 1], elev=45, azim=135)
plt.cla()
axes.scatter(X[:, 0], X[:, 1], X[:, 2], c=y,
               cmap=plt.cm.autumn_r)

axes.set_title("3D Chart")
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_zlabel("Z")

plt.show()
```
