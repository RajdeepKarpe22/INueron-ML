__In this assignment students have to compress racoon grey scale image into<br>
5 clusters. In the end, visualize both raw and compressed image and look<br>
for quality difference.__

```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
import matplotlib.pyplot as plt
import scipy
```

__Fetch image and visualize image__

```python
image = scipy.misc.face(gray=True)
plt.figure(figsize=(10,6))
plt.imshow(image, plt.cm.gray)
plt.show()
```

```python
image.shape
```

```python
rows = image.shape[0]
columns = image.shape[1]
```

```python
image = image.reshape(rows*columns, 1)
```

__Making clusters__

```python
from sklearn.cluster import KMeans

Kmeans = KMeans(n_clusters = 5)
Kmeans.fit(image)
```

```python
clusters = np.asarray(Kmeans.cluster_centers_)
```

```python
clusters
```

```python
labels = np.asarray(Kmeans.labels_)
```

```python
labels
```

```python
labels = labels.reshape(rows, columns)
```

```python
labels
```

```python
plt.imsave("compressed_image.png", labels)
```

```python
image = plt.imread("compressed_image.png")
plt.figure(figsize=(10,6))
plt.imshow(image)
plt.show()
```
